```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('e://maoyan.csv',encoding='gbk')
```

```python
data.head()
```

```python
data['long_time'] = data['long_time'].str.strip('分钟')#去掉分钟
```

```python
data.head()
```

```python
data.info()#一定要查看数据类型！object类型就是String类型的串
```

```python
data['time'] = data['time'].str.strip('上映时间：')#去掉上映时间
data
```

```python
data.head()
```

```python
data['year'] = data['time'].str.split('-').str[0]#新增一列year，取time里的字符串的第一个元素，去掉—
data['month'] = data['time'].str.split('-').str[1]
data.head()
```

```python
data['star'] = data['star'].str.strip('主演：')#去掉 主演：
data.head()
```

```python
data.info()
```

```python
data['long_time'] = data['long_time'].astype('int64')#将时长类型变为Int型
data.dtypes
```

```python
data.head()
```

```python
def get_country(i):
    time= i.split('(')
    if len(time) == 1:
        return '中国'
    else:
        county_1 =  time[1].strip(')')
        if county_1 == '中国香港':
            return '中国'
        elif county_1 == '法国戛纳':
            return '法国'
        else:
            return county_1
```

```python
data['country'] = data['time'].map(get_country)
data
```

```python
data['country'].unique()#唯一值
```

```python
data.head()
```

```python
year_count = data.groupby('year')['year'].count()#年份计数
year_count
```

```python
str = ''
for i in range(50):
    str=str+ ','+data.iloc[i,4]#用坐标进行数据选取
str
```

```python
style_1 = str.split(',')
style = style_1[1:]
style
```

```python
style.count('爱情')
```

```python
from collections import Counter
c = Counter(style)
c
```

```python
style_count = c.most_common(6)#最常见的6种类型
style_count
```

```python
v = []
attr = []
for i in style_count:
    v.append(i[1])#在列表末尾添加新的对象
    attr.append(i[0])
print(v,attr)
```

```python
from pyecharts import Pie
pie = Pie("电影类型分布", title_pos='center', width=900)
pie.add(
    "",
    attr,
    v,
    is_random=True,
    radius=[30, 75],
    rosetype="area",
    is_legend_show=False,
    is_label_show=True,
)
pie.render('e:d.html')
```

```python
#同样处理主演
str_1 = ''
for i in range(50):
    str_1 = str_1 + ',' + data.iloc[i,1]
str_1
```

```python
actor_1 = str_1.split(',')
actor = actor_1[1:]
actor
```

```python
from collections import Counter
a = Counter(actor)
a
```

```python
actor_count = a.most_common(6)
actor_count
```

```python
v = []
attr = []
for i in actor_count:
    v.append(i[1])
    attr.append(i[0])
print(v,attr)
```

```python
from pyecharts import Bar
bar = Bar("")
bar.add("演员", attr, v)
bar.render('e:e.html')
```
